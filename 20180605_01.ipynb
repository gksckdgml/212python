```python
def double(x):
    return x * 2

def triple(x):
    return x * 3
```

```python
def check_number(x):
    if x > 0:
        return "양수"
    elif x == 0:
        return "0"
    else:
        return "음수"
```

```python
print("100 : ",check_number(100), ", 0 : ", check_number(0), ", -1 : ", check_number(-1))
```

```python
def c_num(x):
    return "0" if x == 0 else ( "양수" if x > 0 else "음수")
```

```python
print(c_num(100))
print(c_num(0))
print(c_num(-1))
```
